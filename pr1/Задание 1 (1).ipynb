Задание 7.

На уровне значимости 0,05 проверить статистическую гипотезу о равенстве средней стоимости страховки 13500:
$$H_0:M(X)=13500$$
$$H_1:M(X)\ne13500$$

```python
from scipy.stats import ttest_1samp

t_stat, p_val = ttest_1samp(df['Стоимость'], 13500)

print(f'T-Statistic: {t_stat}')
print(f'p-value: {p_val}')
```

Задание 9.

Оценить влияние пола клиента на стоимость страховки.

```python
# Boxplot for comparing the cost of insurance based on the gender
plt.figure(figsize=(6, 6))
plt.boxplot([df[df['Пол'] == 'М']['Стоимость'], df[df['Пол'] == 'Ж']['Стоимость']], labels=['Male', 'Female'])
plt.title('Boxplot of insurance cost by gender')
plt.ylabel('Cost')
plt.grid(True)
plt.show()

# Calculate the average cost of insurance for men and women
average_cost_male = df[df['Пол'] == 'М']['Стоимость'].mean()
average_cost_female = df[df['Пол'] == 'Ж']['Стоимость'].mean()

average_cost_male, average_cost_female
```

Првести визуальный сравнительный анализ распределений стоимости страховки.

```python
# Group data by gender
df_grouped_by_gender = df.groupby('Пол')['Стоимость']

# Get data for males and females
male_costs = df_grouped_by_gender.get_group('М')
female_costs = df_grouped_by_gender.get_group('Ж')

# Plot histograms
plt.hist(male_costs, alpha=0.5, label='Male', bins=30)
plt.hist(female_costs, alpha=0.5, label='Female', bins=30)

plt.title('Distribution of insurance cost for different genders')
plt.xlabel('Insurance cost')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

Проверить статистичекую гипотезу о равенстве средних значений стоимости страховки для мужчин и женщин на уровне значимости $\alpha=0,05$

$$H_0:M(X)=M(Y)$$
$$H_1:M(X)\ne{M(Y)}$$

```python
from scipy.stats import ttest_ind

# Independent t-test
male_costs = df[df['Пол'] == 'М']['Стоимость']
female_costs = df[df['Пол'] == 'Ж']['Стоимость']

t_stat, p_val = ttest_ind(male_costs, female_costs)

print(f'T-Statistic: {t_stat}')
print(f'p-value: {p_val}')
```

Задание 10.

Вычислить коэффициент точечно-бисериальной корреляции между полом клиента и соимостью страховки.


```python
from scipy.stats import pointbiserialr

# Replace the 'Пол' column with a boolean representation
df_copy = df.copy()
df_copy['Пол'] = df_copy['Пол'].replace({'М': 0, 'Ж': 1})

# Calculate the point-biserial correlation
pb_corr, p_val = pointbiserialr(df_copy['Пол'], df_copy['Стоимость'])

pb_corr, p_val
```

Оцениванить степень зависимости стоимости страховки от индекса массы для курящих и не курящих клиентов путем вычисления коэффициента корреляции Спирмена.
