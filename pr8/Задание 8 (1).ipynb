# **Классификация**

Сгенерировать исходные данные для обучения модели машинного обучения.

Значения параметров выбираются из таблицы 1 "Варианты".docx в соответствие с номером по списку группы в СДО.

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples = 265,
                           n_features = 8,
                           n_classes = 2,
                           n_informative = 8,
                           n_redundant = 0,
                           class_sep = 0.35,
                           random_state = 20)
```

Разделить выборку на обучающую и тестовую.

Объем тестовой выборки 0,2.

Использовать параметр `random_state` в соответствие со своим вариантом.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

Провести сравнительный анализ качества следующих моделей:
- логистическая регрессия;
- решающие деревья;
- бэггинг;
- случайный лес.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Define models
log_reg = LogisticRegression(random_state=20)
dec_tree = DecisionTreeClassifier(random_state=20)
bagging = BaggingClassifier(random_state=20)
random_forest = RandomForestClassifier(random_state=20)
gbc = GradientBoostingClassifier(random_state=20)
svm = SVC(random_state=20)
knn = KNeighborsClassifier()

models = [log_reg, dec_tree, bagging, random_forest, gbc, svm, knn]
model_names = ['Logistic Regression', 'Decision Tree', 'Bagging', 'Random Forest', 'Gradient Boosting', 'SVM', 'KNN']

for model, name in zip(models, model_names):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy of {name}: {acc}")
```

Выполнить следующие этапы анализа:
- определить оптимальные гиперпараметры модели;
- обучить модель с найденными оптимальными гиперпараметрами;
- оценить качество модели на обучающей и тестовой выборках:
 - метрика accuracy;
 - матрица ошибок;
 - отчет по метрикам качества;
 - ROC-кривая.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt

# Define parameters grids
params_grid_log_reg = {'penalty': ['l1', 'l2', 'elasticnet', 'none'], 'C': [0.01, 0.1, 1.0, 10, 100]}
params_grid_dec_tree = {'criterion': ['gini', 'entropy'], 'max_depth': [2,5,10,20]}
params_grid_bagging = {'n_estimators': [10, 50, 100, 200], 'max_samples': [0.5, 1.0]}
params_grid_random_forest = {'n_estimators': [10, 50, 100, 200], 'max_depth': [2,5,10,20], 'criterion':['gini', 'entropy']}
params_grid_svm = {'C': [0.1, 1, 10, 100], 'gamma': [0.1, 0.01, 0.001, 0.0001], 'kernel': ['linear', 'poly', 'rbf', 'sigmoid'], 'probability': [True]}
params_grid_knn = {'n_neighbors': list(range(1,31)), 'metric': ['minkowski', 'manhattan', 'euclidean'], 'weights': ['uniform', 'distance']}
params_grid_gbc = {'n_estimators': [10, 50, 100, 200], 'learning_rate': [0.001, 0.01, 0.1, 1], 'max_depth': [2,3,4,5]}

params_grids = [params_grid_log_reg, params_grid_dec_tree, params_grid_bagging, params_grid_random_forest, params_grid_svm, params_grid_knn, params_grid_gbc]

models = [log_reg, dec_tree, bagging, random_forest, svm, knn, gbc]
model_names = ['Logistic Regression', 'Decision Tree', 'Bagging', 'Random Forest', 'SVM', 'KNN', 'GradientBoosting']

for model, name, params_grid in zip(models, model_names, params_grids):
    # GridSearchCV to determine optimal parameters
    grid = GridSearchCV(model, params_grid, cv=5)
    grid.fit(X_train, y_train)
    print(f"Model: {name}\\nBest parameters: {grid.best_params_}")

    # Re-train the model with optimal parameters
    model.set_params(**grid.best_params_)
    model.fit(X_train, y_train)
    
    # Predict and evaluate the model
    y_pred = model.predict(X_test)
    y_score = model.predict_proba(X_test)[:,1]

    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    print(f'Accuracy: {acc}')

    # Confusion Matrix
    conf_m = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix:')
    print(conf_m)

    # Classification Report
    report = classification_report(y_test, y_pred)
    print('Classification Report:')
    print(report)

    # ROC Curve
    fpr, tpr, thresholds = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=1, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()
```

Результаты сравнительного анализа оформить в виде таблицы, пример которой приведен ниже.

```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import precision_recall_curve, auc

# Create lists to save metrics
accuracy_list = []
precision_list = []
recall_list = []
f1_list = []
roc_auc_list = []
pr_auc_list = []

# Calculate and save metrics
for model, name, params_grid in zip(models, model_names, params_grids):
    grid = GridSearchCV(model, params_grid, cv=5)
    grid.fit(X_train, y_train)
    
    y_pred = grid.predict(X_test)
    y_score = grid.predict_proba(X_test)[:,1]
    
    accuracy_list.append(accuracy_score(y_test, y_pred))
    precision_list.append(precision_score(y_test, y_pred))
    recall_list.append(recall_score(y_test, y_pred))
    f1_list.append(f1_score(y_test, y_pred))
    roc_auc_list.append(roc_auc_score(y_test, y_score))
    
    precision, recall, _ = precision_recall_curve(y_test, y_score)
    pr_auc_list.append(auc(recall, precision))

# Create and display the DataFrame
metrics_df = pd.DataFrame({
    'Model': model_names, 
    'Accuracy': accuracy_list, 
    'Precision': precision_list, 
    'Recall': recall_list, 
    'F1': f1_list, 
    'ROC AUC': roc_auc_list, 
    'PR AUC': pr_auc_list
})

metrics_df
```

```python
metrics_df = metrics_df.sort_values(by=list(metrics_df.columns[1:]), ascending=False)
metrics_df
```

Обосновать выбор лучшей модели.

Оценить важность признаков на основе выбранной модели.

```python
# As this task requires a subjective judgement, it's not possible to provide a concrete code solution. 
# However, here's a general guideline on how this can be achieved.

# 1. To justify the choice of the best model, one can use the following template:
best_model = models[accuracy_list.index(max(accuracy_list))]
print(f"The best model is {type(best_model).__name__} due to its highest accuracy score in comparison to other models.")

# 2. To evaluate feature importance, it's often model-dependent. But for tree-based models (DecisionTree, RandomForest), you can do the following:
if hasattr(best_model, 'feature_importances_'):
    feature_importance = best_model.feature_importances_
    print(f"Feature importance of the best model: {feature_importance}")
else:
    print(f"The model {type(best_model).__name__} does not provide feature importances.")
```

# **Регрессия**

Сгенерировать исходные данные для обучения модели машинного обучения.

Значения параметров выбираются из таблицы 2 "Варианты".docx в соответствие с номером по списку группы в СДО.

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 265,
                       n_features = 8,
                       noise = 1.2,
                       n_informative = 6,
                       random_state = 20)
```

Разделить выборку на обучающую и тестовую.

Объем тестовой выборки 0,2.

Использовать параметр `random_state` в соответствие со своим вариантом.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

Провести сравнительный анализ качества следующих моделей:
- линейная регрессионная модель;
- решающие деревья;
- бэггинг;
- случайный лес.

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error

# Define models
lin_reg = LinearRegression()
dec_tree_reg = DecisionTreeRegressor(random_state=20)
bagging_reg = BaggingRegressor(random_state=20)
random_forest_reg = RandomForestRegressor(random_state=20)

models_reg = [lin_reg, dec_tree_reg, bagging_reg, random_forest_reg]
model_names_reg = ['Linear Regression', 'Decision Tree Regressor', 'Bagging Regressor', 'Random Forest Regressor']

for model_reg, name_reg in zip(models_reg, model_names_reg):
    model_reg.fit(X_train, y_train)
    y_pred_reg = model_reg.predict(X_test)
    mse = mean_squared_error(y_test, y_pred_reg)
    mape = mean_absolute_percentage_error(y_test, y_pred_reg)
    r2 = r2_score(y_test, y_pred_reg)
    
    print(f"Metrics of {name_reg}: Mean Squared Error: {mse}, Mean Absolute Percentage Error: {mape}, R2 Score: {r2}")
```

Выполнить следующие этапы анализа:
- определить оптимальные гиперпараметры модели;
- обучить модель с найденными оптимальными гиперпараметрами;
- оценить качество модели на обучающей и тестовой выборках:
 - MSE;
 - MAPE;
 - R2.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

# Define parameters grids for regression models
params_grid_lin_reg = {'fit_intercept': [True, False], 'normalize': [True, False]}
params_grid_dec_tree_reg = {'criterion': ['mse', 'friedman_mse', 'mae', 'poisson'], 'max_depth': [2,5,10,20]}
params_grid_bagging_reg = {'n_estimators': [10, 50, 100, 200], 'max_samples': [0.5, 1.0]}
params_grid_random_forest_reg = {'n_estimators': [10, 50, 100, 200], 'max_depth': [2,5,10,20], 'criterion':['mse', 'mae']}

params_grids_reg = [params_grid_lin_reg, params_grid_dec_tree_reg, params_grid_bagging_reg, params_grid_random_forest_reg]

models_reg = [lin_reg, dec_tree_reg, bagging_reg, random_forest_reg]
model_names_reg = ['Linear Regression', 'Decision Tree Regressor', 'Bagging Regressor', 'Random Forest Regressor']

# Perform model selection, training, and evaluation
for model_reg, name_reg, params_grid_reg in zip(models_reg, model_names_reg, params_grids_reg):
    # GridSearchCV to determine optimal parameters
    grid_reg = GridSearchCV(model_reg, params_grid_reg, cv=5)
    grid_reg.fit(X_train, y_train)

    # Print best parameters
    print(f"Best parameters for {name_reg}: {grid_reg.best_params_}")

    # Re-train the model with optimal parameters
    model_reg.set_params(**grid_reg.best_params_)
    model_reg.fit(X_train, y_train)

    # Predict and calculate metrics with the model
    y_pred_reg = model_reg.predict(X_test)
    mse = mean_squared_error(y_test, y_pred_reg)
    mape = mean_absolute_percentage_error(y_test, y_pred_reg)
    r2 = r2_score(y_test, y_pred_reg)

    # Print metrics
    print(f"Metrics for model {name_reg}:\nMean Squared Error: {mse}\nMean Absolute Percentage Error: {mape}\nR2 Score: {r2}\n")
```

Результаты сравнительного анализа оформить в виде таблицы по аналогии с таблицей для задачи классификации.

```python
# Create lists to save metrics
mse_list = []
mape_list = []
r2_list = []

# Calculate and save metrics
for model_reg, name_reg, params_grid_reg in zip(models_reg, model_names_reg, params_grids_reg):
    grid_reg = GridSearchCV(model_reg, params_grid_reg, cv=5)
    grid_reg.fit(X_train, y_train)
    
    y_pred_reg = grid_reg.predict(X_test)
    
    mse_list.append(mean_squared_error(y_test, y_pred_reg))
    mape_list.append(mean_absolute_percentage_error(y_test, y_pred_reg))
    r2_list.append(r2_score(y_test, y_pred_reg))

# Create and display the DataFrame
metrics_df_reg = pd.DataFrame({
    'Model': model_names_reg, 
    'MSE': mse_list, 
    'MAPE': mape_list, 
    'R2': r2_list, 
})

metrics_df_reg
```

```python
metrics_df_reg = metrics_df_reg.sort_values(by='R2', ascending=False)
metrics_df_reg
```

Обосновать выбор лучшей модели.

Реализовать прогнозирование целевого признака для нескольких произвольных объектов, которые не участвовали в обучении модели.

```python
# Justifying the choice of the best model.
# The best model is chosen based on the highest R2 score
best_model_reg = models_reg[r2_list.index(max(r2_list))]
print(f"The best model is {type(best_model_reg).__name__} due to its highest R2 score in comparison to other models.")

# Predicting the target feature for new arbitrary data points that weren't part of the training set.
# Assuming new_data is a list of lists, where each list contains 8 feature values
new_data = [[0.1, 1.2, 3.1, 4.5, 5.7, 6.0, 7.8, 8.3], [0.9, 1.0, 1.1, 1.2, 2.3, 2.4, 2.5, 3.6]] 
new_data_pred_reg = best_model_reg.predict(new_data)
print(new_data_pred_reg)
```
